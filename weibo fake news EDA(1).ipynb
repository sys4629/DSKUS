```python
num_df.boxplot()
```

### After Outlier handling

```python
Q1 = num_df[['like_num','repost_num']].quantile(q=0.25)
Q3 = num_df[['like_num','repost_num']].quantile(q=0.75)
print(Q1)
print(Q3)
```

```python
IQR = Q3-Q1
IQR
```

```python
IQR_df = num_df[(num_df['like_num'] <= Q3['like_num']+1.5*IQR['like_num']) & (num_df['like_num'] >= Q1['like_num']-1.5*IQR['like_num'])]
IQR_df = IQR_df[(IQR_df['repost_num'] <= Q3['repost_num']+1.5*IQR['repost_num']) & (IQR_df['repost_num'] >= Q1['repost_num']-1.5*IQR['repost_num'])]
IQR_df = IQR_df[['like_num','repost_num']]
print('original :', len(num_df))
print('IQR_df :', len(IQR_df))
```

```python
IQR_df.boxplot()
```

# natural language data

```python
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
nlp_df = pd.DataFrame()
# Character counts
nlp_df["char_count"] = df["text"].map(lambda x: len(x))
# Word counts
nlp_df["word_count"] = df["text"].map(lambda x: len(x.split()))
# Sentence counts
nlp_df["sent_count"] = df["text"].map(lambda x: len(sent_tokenize(x)))
# Average word length
nlp_df["avg_word_len"] = df["text"].map(lambda x: np.mean([len(w) for w in str(x).split()]))
# Average sentence length
nlp_df["avg_sent_len"] = df["text"].map(lambda x: np.mean([len(w.split()) for w in sent_tokenize(x)]))
nlp_df
```

```python
#set seaborn plotting aesthetics as default
sns.set()

fig, axes = plt.subplots(3, 2)
fig.set_size_inches(12, 12)

#create boxplot in each subplot
sns.kdeplot(nlp_df['char_count'],shade=True,color='Blue', ax=axes[0,0])
sns.kdeplot(nlp_df['word_count'],shade=True,color='Green', ax=axes[0,1])
sns.kdeplot(nlp_df['sent_count'],shade=True,color='Red', ax=axes[1,0])
sns.kdeplot(nlp_df['avg_word_len'],shade=True,color='Orange', ax=axes[1,1])
sns.kdeplot(nlp_df['avg_sent_len'],shade=True,color='Purple', ax=axes[2,0])
plt.show()
```
